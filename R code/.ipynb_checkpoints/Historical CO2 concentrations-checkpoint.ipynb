```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
glacial =  pd.read_csv('carbon.csv', sep=';')
glacial.set_index('age', inplace=True)
glacial.head()
```

```python
glacial_subset =  pd.read_csv('carbon_subsets.csv', sep=';')
glacial_subset.set_index('age', inplace=True)
glacial_subset.head()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes2 = fig.add_axes([0.57,0.65,0.4,0.35])

axes1.plot(glacial)
axes1.set_xlabel('Age (years before present)')
axes1.set_ylabel('CO2 concentration (ppm)')

axes2.plot(glacial_subset)

plt.savefig('Glacial CO2.png')
```

```python
glacial.plot(figsize=(15,10))
#plt.xticks(rotation=60)
#plt.xlim(440000,10)
plt.xlabel('Age (years before present)')
plt.ylabel('CO2 concentration (ppm)')
# Specify the axes
plt.axes([0.18,0.62,0.38,0.25])

# Plot the sliced series in red using the current axes
plt.plot(glacial_subset)
#plt.xlim(210000,178000)
#plt.xticks(rotation=45)
#plt.title('2007/11-2008/04')
plt.show()
plt.savefig('Historical CO2 concentrations')
```
